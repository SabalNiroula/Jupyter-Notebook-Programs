```python
treatment_cpy['hba1c_change'] = treatment_cpy['hba1c_start'] - treatment_cpy['hba1c_end']  
for i in  range(len(treatment_cpy['given_name'])):

    treatment_cpy.loc[i,  'surname'] = treatment_cpy['surname'][i].capitalize()
    treatment_cpy.loc[i,  'given_name'] = treatment_cpy['given_name'][i].capitalize()

treatment_cpy
```

```python
#next problem we want to solve is we want to have treatment column where auralin and novodra will stored
        #source         #columns to leaved unchanged                                                                   #new col with col name       #col name where values are stored of changed columns 
treatment_cpy  = pd.melt(treatment_cpy, id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change'],  var_name = 'treatment', value_name = "dose")
treatment_cpy
```

```python
treatment_cpy = treatment_cpy[treatment_cpy["dose"] != "-"] 
treatment_cpy
```

```python
#since dose is object we convert into string using "str" function
#here expand = True in split returns a DataFrame
# if you get all the warning just use the following code
treatment_cpy._is_copy = False
treatment_cpy[['start','end']] = treatment_cpy['dose'].str.split('-', expand = True)
treatment_cpy
```

```python
treatment_cpy.drop(columns = [ 'dose'], inplace=True)
treatment_cpy = treatment_cpy.drop_duplicates()
treatment_cpy
```

```python
def isNaN(string):
    return string != string
patient["contact"][14]
```

```python
import re
rgx = r'\d'
pattern = re.compile(rgx)
patternS = re.compile(r'[^\d\-\(\)\+]')
contact = []
email = []
for i in range(len(patient)):
    string = patient['contact'][i]
    if(isNaN(string)): 
        contact.append("No Contact")
        email.append("No Email")
        continue
    matchI = pattern.findall(string)
    matchS = patternS.findall(string)
    em = "".join(matchS)
    n = len(matchI)
    con = ''
#     print(matchI)
    if n == 11 :
        con = "(+" + matchI[0] + ") "
        for k in range(1, n):
            con+=matchI[k]
            if (k == 3 or k == 6 ):
                con+="-"
    elif n==10:
        for k in range(n):
            con += matchI[k]
            if (k == 2 or k == 5 ):
                con += "-"
    contact.append(con)
    email.append(em)
```

```python
patient['email'] = email
patient['number'] = contact
patient.drop(columns = ["contact"], inplace = True)
patient
```
