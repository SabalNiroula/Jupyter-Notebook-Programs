```python
#let's find the most destrutive batsman in ipl history
#we are looking for the player who's strike rate=(runs/no of ball) is greatest in 16-20 over and has played more than 200 balls overall
```

```python
delivery = data[data['over']>15]
```

```python
all_batsman = delivery.groupby('batsman')['batsman_runs'].count()
msk = all_batsman>200 #because we want batsman who's played more than 200 balls 
batsman_detail = all_batsman[msk]
batsman_list = batsman_detail[msk].index.tolist() #have all the names of batsman who played more than 200 balls 
count_run = delivery.groupby('batsman')['batsman_runs'].sum() # summing all the runs
count_run = count_run[msk] #masking to the players who have played more than 200 balls
batsman_list
```

```python
def result(batsman):
    balls = batsman_detail[batsman]
    runs = count_run[batsman]
    return (runs/balls)*100
```

```python

sr = (count_run/batsman_detail)*100
sr
```

```python
name =""
max_strike = 0.0
for i in range(len(batsman_list)):
    rate = result(i)
    if (rate>max_strike):
        max_strike = rate
        name = batsman_list[i]
    
print(name, max_strike)
```

```python
#lets do the same thing using isin() function 
#we have the list of all batsman who played more than 200 balls and want to calculate runs and ball played by all 43 batsman
final = delivery[delivery['batsman'].isin(batsman_list)]
batsman_detail.shape
```

```python
balls =  final.groupby('batsman')['batsman_runs'].count()
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
 sr =(runs/balls)*100
```
